Докажем что приведенные примеры порядков являются отношением порядка на мономах
Для этого проверим аксиомы отношения порядка 
1) Транзитивность:
    A ≺ B и B ≺ C ⇒ A ≺ C
2) Линейность:
    ∀A, B, A = B или A ≺ B или B ≺ A
3) Совместимость:
    A ≺ B ⇒ A * C ≺ B * C
# Лексикографический порядок (lex):
    1) B ≺ A и B ≺ C
       Значит есть позиция монома где B ≺ A и C ≺ B. Тогда возьмем самую левую такую позицию, на ней будет верно, что C ≺ A
    2) Да, тк два монома либо равны, либо
       на какой то позиции будут различаться, значит выполнится одно из условий линейности
    3) B ≺ A 
       Значит существует позиция, где они различаются. При умножении на С разница в этой позиции останется. Значит аксиома выполняется
    Лексикографический порядок является отношением
# Порядок по степени (grlex):
    1) B ≺ A и B ≺ C
       Если степени мономов разные, то C ≺ A
       Если степени одинаковые доказана транзитивность lex
    2) Если степени разные, то по
       определению будет одно из отношений 
       B ≺ A или A ≺ B
       Если степени равны, то мы уже доказывали линейность lex
    3) B ≺ A
       При умножении на C степень исходного
       монома увеличится на степень C. 
       Если |B| ≺ |A| то
       |B| + |C| ≺ |A| + |C| а значит B * C ≺ A * C
       Если |A| = |B| и B ≺(lex) A
       по совместимости lex верно B * C ≺ A * C
    Порядок степени является отношением



# Редукция 
f → f - (LT(f)/LT(gᵢ))·gᵢ
LT(gᵢ) должно делить LT(f)

## Если получился 0?
  
Полином f полностью выражается через порождающие g₁, ..., gₘ, то есть  
f = h₁g₁ + ... + hₘgₘ  
для некоторых hᵢ ∈ K[x₁, ..., xₙ].
f ∈ I = (g₁, ..., gₘ)

## Если получился не 0?

Полином f не принадлежит идеалу I. Остаток r — нередуцируемая часть f.
f ∉ I

## Существует ли f ∈ I, который не редуцируется к 0?

Да, если G = {g₁, ..., gₘ} не является базисом Грёбнера.
- Существуют f ∈ I
- Которые не редуцируются к 0
- Потому что для их представления через gᵢ нужны МЛАДШИЕ члены
- А редукция работает только со СТАРШИМИ членами

## Зависит ли результат от порядка применения gᵢ?
Да, зависит.
## Пример:

I = (g₁, g₂) где:
- g₁ = x² - y
- g₂ = xy - 1

f = x²y + xy - x - y

---

## Сначала g₁, потом g₂

LT(f) = x²y
- LM(g₁) = x² делит x²y 
- Редуцируем: f → f - (x²y/x²)·g₁  
  = (x²y + xy - x - y) - y·(x² - y)  
  = x²y + xy - x - y - x²y + y²  
  = xy - x - y + y²
LT = xy  
- LM(g₂) = xy делит xy 
- Редуцируем: → (xy - x - y + y²) - (xy/xy)·g₂  
  = xy - x - y + y² - (xy - 1)  
  = xy - x - y + y² - xy + 1  
  = -x - y + y² + 1

**Результат**: r₁ = -x - y + y² + 1

---

## Сначала g₂, потом g₁

LT(f) = x²y
- LM(g₂) = xy делит x²y 
- Редуцируем: f → f - (x²y/xy)·g₂  
  = (x²y + xy - x - y) - x·(xy - 1)  
  = x²y + xy - x - y - x²y + x  
  = xy - y

LT = xy
- LM(g₂) = xy делит xy 
- Редуцируем: → (xy - y) - (xy/xy)·g₂  
  = xy - y - (xy - 1)  
  = xy - y - xy + 1 = -y + 1

**Результат**: r₂ = -y + 1

r₁ ≠ r₂

## Обоснование 
На каждом шаге есть выбор, через какой gᵢ редуцировать, и этот выбор влияет на промежуточные результаты и конечный остаток.

# Любой базис Гребнера можно привести к редуцированному 
## 1. Нормируем ведущие коэффициенты 
    Очевидно, умножаем моном на коэффициент, обратный к LC(gᵢ)
    На данном этапе мы не нарушаем LM(f) делится на LM(gᵢ)
    тк LM(gᵢ) не меняется
## 2. Удаление избыточных
    Пока существуют gᵢ и gⱼ, такие что LM(gᵢ) делит LM(gⱼ) удаляем gⱼ 
    Получатся взаимнопростые мономы 
    gⱼ избыточен - его можно выразить через другие элементы
    Удаление gⱼ не меняет идеал I потому что gⱼ уже является линейной комбинацией других элементов базиса, а значит свойство базиса Гренберга не нарушается
## 3. G нормальность
    Берем полином gᵢ
    Ищем моном m в gᵢ такой, что m делится на LM(gⱼ)
    Выполняем операцию редукции:
        gᵢ → gᵢ - (коэфф(m) * m / LM(gⱼ)) * gⱼ
    Эта операция устраняет моном m из gᵢ
    Повторяем пока все кратные не устранятся 

    Идеал I не меняется: Каждая редукци является просто добавлением или вычитанием элемента из идеала I 

    Свойство базиса Грёбнера сохраняется: Ведущий моном LM(gᵢ) никогда не меняется в процессе редукции, потому что:
        LM(gᵢ) не имеет кратных
        Удаляем только младшие мономы m
