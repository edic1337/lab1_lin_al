```python
import time
from typing import List
import math
from sympy import totient as sympy_euler_phi

def gcd(a: int, b: int) -> int:
    return math.gcd(a, b)

def euler_phi_direct(n: int) -> int: # прямой перебор 
    count = 0
    for i in range(1, n + 1):
        if gcd(i, n) == 1:
            count += 1
    return count 

def euler_phi_factor(n: int) -> int: # разложение
    if n == 1:
        return 1
    def factorize(n: int) -> Dict[int, int]:
        factors = {}
        # проверяем делимость на 2
        while n % 2 == 0:
            factors[2] = factors.get(2, 0) + 1
            n //= 2
        # проверяем нечётные делители
        factor = 3
        while factor * factor <= n:
            while n % factor == 0:
                factors[factor] = factors.get(factor, 0) + 1
                n //= factor
            factor += 2
        # если остаток простой
        if n > 1:
            factors[n] = factors.get(n, 0) + 1
        return factors
    fact = factorize(n)
    res = n
    for multiplier in fact:
        res = res * (multiplier - 1) // multiplier 
    return res

test_values = [10, 100, 1000, 5000, 10000, 50000]

def compare_euler_phi_methods(test_values: List[int]) -> dict: # сравнение времени
    print("сравнение")
    print(f"{'n':<8} {'прямой (с)':<12} {'разложение (с)':<12} {'SymPy (с)':<12}")
    print("-" * 70)
    for n in test_values:
        start = time.time() # прямой перебор
        phi1 = euler_phi_direct(n)
        time1 = time.time() - start
        start = time.time() # через разложение
        phi2 = euler_phi_factor(n)
        time2 = time.time() - start
        start = time.time() # SymPy
        phi3 = sympy_euler_phi(n)
        time3 = time.time() - start
        print(f"{n:<8} {time1:<12.6f} {time2:<12.6f} {time3:<12.6f}")
compare_euler_phi_methods(test_values)
```

```python
import sage.all
from sage.rings.finite_rings.finite_field_constructor import GF

N = 501443 % 20

def extended_gcd_poly(a, b):
    if b == 0:
        return (1, 0, a)
    x0, x1 = 1, 0
    y0, y1 = 0, 1
    r0, r1 = a, b
    while r1 != 0:
        q = r0 // r1
        r2 = r0 - q * r1
        x2 = x0 - q * x1
        y2 = y0 - q * y1
        r0, r1 = r1, r2
        x0, x1 = x1, x2
        y0, y1 = y1, y2
    return (x0, y0, r0)

def find_gcd_representation():
    print("нахождение НОД и его линейного представления")
    F11 = GF(11)
    R11 = F11['x']
    x = R11.gen()
    r_coeffs = [] # коэффициенты r_i = (i + N) % 11
    for i in range(8):
        value = (i + N) % 11
        r_coeffs.append(F11(value)) 
    s_coeffs = [] # коэффициенты s_i = (i + N) % 11
    for i in range(4):
        value = (i + N) % 11
        s_coeffs.append(F11(value))
    f = r_coeffs[0] + r_coeffs[1]*x + r_coeffs[2]*x**2 + r_coeffs[3]*x**3 + r_coeffs[4]*x**4 + r_coeffs[5]*x**5 + r_coeffs[6]*x**6 + r_coeffs[7]*x**7
    g = s_coeffs[0] + s_coeffs[1]*x + s_coeffs[2]*x**2 + s_coeffs[3]*x**3
    print(f"f(x) = {f}")
    print(f"g(x) = {g}")
    gcd_fg = f.gcd(g) # находим НОД
    print(f"НОД(f,g) = {gcd_fg}")
    u, v, w = extended_gcd_poly(f, g) # находим линейное представление
    inv_w = w**-1 # нормализуем результат
    u_normalized = u * inv_w
    v_normalized = v * inv_w
    print(f"u(x) = {u_normalized}")
    print(f"v(x) = {v_normalized}")
    check = u_normalized*f + v_normalized*g # проверка
    print(f"u*f + v*g = {check}")

find_gcd_representation()
```

```python
import sage.all
from sage.rings.finite_rings.finite_field_constructor import GF

N = 501443 % 20

def find_inverse():
    print("нахождение обратного элемента f^{-1} mod g")
    F13 = GF(13)
    R13 = F13['x']
    x = R13.gen()
    s_coeffs = [] # коэффициенты s_i = (i + N) % 11
    for i in range(3):
        value = (i + N) % 11
        s_coeffs.append(F13(value))
    # строим полиномы
    f = s_coeffs[2]*x**2 + s_coeffs[1]*x + s_coeffs[0]
    g = x**8 + x**4 + x**3 + 6*x + 2
    print(f"f(x) = {f}")
    print(f"g(x) = {g}")
    gcd_val, h, _ = f.xgcd(g) # находим обратный элемент
    if gcd_val == 1: # проверяем, что обратный существует
        print(f"f^(-1) mod g = {h}")
        check = (f * h) % g # проверка
        print(f"проверка: f * h mod g = {check}")
    else:
        print("обратный элемент не существует")

find_inverse()
```

```python
import sage.all
from sage.rings.finite_rings.finite_field_constructor import GF

def generate_irreducible_polynomials(q: int, d: int) -> list:
    Fq = GF(q) # конечное поле F_q
    R = Fq['x'] # кольцо полиномов над F_q
    x = R.gen() # переменная для верного построения 
    irreducible_polys = []
    # перебираем все возможные коэффициенты
    from itertools import product
    for coeffs in product(Fq, repeat=d+1): # декартово произведение, где каждый элемент длинной repeat
        if coeffs[d] != 0:  # старший коэффициент не должен быть 0
            # строим полином обычным циклом
            poly = Fq(0)
            for i in range(d+1):
                poly = poly + coeffs[i] * (x**i)
            if poly.is_irreducible():
                irreducible_polys.append(poly)
    return irreducible_polys

test_cases = [(2,2), (2,3), (2,4), (3,2), (3,3), (5,2)]

for q, d in test_cases:
    print(f"\nнеприводимые полиномы степени {d} над F_{q}:")
    polys = generate_irreducible_polynomials(q, d)
    for poly in polys:
        print(f"  {poly}")
    print(f"всего: {len(polys)}")
```
