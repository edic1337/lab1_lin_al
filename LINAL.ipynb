```python
from typing import List

def palindromic_squares_and_circular_primes() -> tuple[List[int], List[int]]:
    palindrom_a = []
    palindrom_p = []
    
    def is_prime(n: int) -> bool: # –ø—Ä–æ—Å—Ç–æ—Ç–∞ 
        if n < 2:
            return False
        if n == 2:
            return True
        if n % 2 == 0:
            return False
        for i in range(3, int(n**0.5) + 1, 2):
            if n % i == 0:
                return False
        return True
    
    def cyclic(n: int) -> List[int]: # —Ü–∏–∫–ª–∏—á–µ—Å–∫–∏–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏
        s = str(n)
        permutations = []
        for i in range(len(s)):
            permutation = int(s[i:] + s[:i])
            permutations.append(permutation)
        return permutations
    for i in range(1, 100000):
        a = str(i)
        s = str(i**2)
        if a == a[::-1] and s == s[::-1]:
            palindrom_a.append(i)  # –ø–∞–ª–∏–Ω–¥—Ä–æ–º—ã( –∞ –∏ –∞^2)
    for k in range(2, 1000000):
        if is_prime(k):
            permutation = cyclic(k)
            all_prime = True
            for perm in permutation:
                if not is_prime(perm):
                    all_prime = False
                    break
            if all_prime:
                palindrom_p.append(k) # –ø—Ä–æ—Å—Ç—ã–µ –∏ –∏—Ö –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏

    return palindrom_a, palindrom_p
    
```

```python
from typing import Dict
import math

def factorial_plus_one_factors() -> Dict[int, Dict[int, int]]:
    
    def is_prime(n: int) -> bool: # –ø—Ä–æ—Å—Ç–æ—Ç–∞ 
        if n < 2:
            return False
        if n == 2:
            return True
        if n % 2 == 0:
            return False
        for i in range(3, int(n**0.5) + 1, 2):
            if n % i == 0:
                return False
        return True
    
    def factorize(n: int) -> Dict[int, int]:
        factors = {}
        # –ø—Ä–æ–≤–µ—Ä—è–µ–º –¥–µ–ª–∏–º–æ—Å—Ç—å –Ω–∞ 2
        while n % 2 == 0:
            factors[2] = factors.get(2, 0) + 1
            n //= 2
        # –ø—Ä–æ–≤–µ—Ä—è–µ–º –Ω–µ—á—ë—Ç–Ω—ã–µ –¥–µ–ª–∏—Ç–µ–ª–∏
        factor = 3
        while factor * factor <= n:
            while n % factor == 0:
                factors[factor] = factors.get(factor, 0) + 1
                n //= factor
            factor += 2
        # –µ—Å–ª–∏ –æ—Å—Ç–∞—Ç–æ–∫ –ø—Ä–æ—Å—Ç–æ–π
        if n > 1:
            factors[n] = factors.get(n, 0) + 1
        return factors
    results = {}
    fact = 1
    for n in range(2, 51):
        fact *= n
        results[n] = factorize(fact + 1) # results[4] = {5: 2} 
    return results
    # –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –ø—Ä–æ—Å—Ç—ã—Ö –¥–µ–ª–∏—Ç–µ–ª–µ–π: 12 (–ø—Ä–∏ n=50)
    # –±–æ–ª—å—à–∏–µ –ø—Ä–æ—Å—Ç—ã–µ –º–Ω–æ–∂–∏—Ç–µ–ª–∏ (>10‚Å∂): n=13: 1713311273364631, n=22: 93799610095769647

factorial_plus_one_factors()
```

```python
import time
from typing import List
import math
from sympy import totient as sympy_euler_phi

def gcd(a: int, b: int) -> int:
    return math.gcd(a, b)

def euler_phi_direct(n: int) -> int: # –ø—Ä—è–º–æ–π –ø–µ—Ä–µ–±–æ—Ä 
    count = 0
    for i in range(1, n + 1):
        if gcd(i, n) == 1:
            count += 1
    return count 

def euler_phi_factor(n: int) -> int: # —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ
    if n == 1:
        return 1
    def factorize(n: int) -> Dict[int, int]:
        factors = {}
        # –ø—Ä–æ–≤–µ—Ä—è–µ–º –¥–µ–ª–∏–º–æ—Å—Ç—å –Ω–∞ 2
        while n % 2 == 0:
            factors[2] = factors.get(2, 0) + 1
            n //= 2
        # –ø—Ä–æ–≤–µ—Ä—è–µ–º –Ω–µ—á—ë—Ç–Ω—ã–µ –¥–µ–ª–∏—Ç–µ–ª–∏
        factor = 3
        while factor * factor <= n:
            while n % factor == 0:
                factors[factor] = factors.get(factor, 0) + 1
                n //= factor
            factor += 2
        # –µ—Å–ª–∏ –æ—Å—Ç–∞—Ç–æ–∫ –ø—Ä–æ—Å—Ç–æ–π
        if n > 1:
            factors[n] = factors.get(n, 0) + 1
        return factors
    fact = factorize(n)
    res = n
    for multiplier in fact:
        res = res * (multiplier - 1) // multiplier 
    return res

test_values = [10, 100, 1000, 5000, 10000, 50000]

def compare_euler_phi_methods(test_values: List[int]) -> dict: # —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–∏
    print("—Å—Ä–∞–≤–Ω–µ–Ω–∏–µ")
    print(f"{'n':<8} {'–ø—Ä—è–º–æ–π (—Å)':<12} {'—Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ (—Å)':<12} {'SymPy (—Å)':<12}")
    print("-" * 70)
    for n in test_values:
        start = time.time() # –ø—Ä—è–º–æ–π –ø–µ—Ä–µ–±–æ—Ä
        phi1 = euler_phi_direct(n)
        time1 = time.time() - start
        start = time.time() # —á–µ—Ä–µ–∑ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ
        phi2 = euler_phi_factor(n)
        time2 = time.time() - start
        start = time.time() # SymPy
        phi3 = sympy_euler_phi(n)
        time3 = time.time() - start
        print(f"{n:<8} {time1:<12.6f} {time2:<12.6f} {time3:<12.6f}")
compare_euler_phi_methods(test_values)
```

```python
import sage.all
from sage.rings.finite_rings.integer_mod_ring import Integers
from sage.groups.perm_gps.permgroup_named import SymmetricGroup

N = 501443 % 20 
m = 4 + N % 5
t = 12
elem_numb = t % m

def isomorphism_of_cyclic_subgroup_Zm_star(N: int) -> dict:
    Z_m = Integers(m)
    # –ø—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ —ç–ª–µ–º–µ–Ω—Ç –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –º—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω–æ–π –≥—Ä—É–ø–ø–µ
    if gcd(elem_numb, m) != 1:
        print(f"–û–®–ò–ë–ö–ê: {elem_numb} –Ω–µ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç Z_{m}*")
        return {}
    t_element = Z_m(elem_numb)
    order_subgroup = t_element.multiplicative_order()  # –ø–æ—Ä—è–¥–æ–∫ —Ü–∏–∫–ª–∏—á–µ—Å–∫–æ–π –ø–æ–¥–≥—Ä—É–ø–ø—ã
    print(f"–º—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω–∞—è –≥—Ä—É–ø–ø–∞ Z_{m}*")
    print(f"—ç–ª–µ–º–µ–Ω—Ç t = {elem_numb} (–∏—Å—Ö–æ–¥–Ω—ã–π {t} mod {m})")
    print(f"–ø–æ—Ä—è–¥–æ–∫ –ø–æ–¥–≥—Ä—É–ø–ø—ã ‚ü®{elem_numb}‚ü©: {order_subgroup}") 
    # —ç–ª–µ–º–µ–Ω—Ç—ã –ø–æ–¥–≥—Ä—É–ø–ø—ã –≤ Z_m*
    subgroup_elements = []
    for i in range(order_subgroup):
        element = t_element ** i
        subgroup_elements.append(element)
    print(f"–≠–ª–µ–º–µ–Ω—Ç—ã –ø–æ–¥–≥—Ä—É–ø–ø—ã: {subgroup_elements}")
    print()
    # —Å–æ–∑–¥–∞–µ–º –∏–∑–æ–º–æ—Ä—Ñ–Ω—É—é —Ü–∏–∫–ª–∏—á–µ—Å–∫—É—é –ø–æ–¥–≥—Ä—É–ø–ø—É –≤ S_d
    d = order_subgroup
    S_d = SymmetricGroup(d)
    # —Ü–∏–∫–ª (1,2,3,...,d)
    cycle_list = []
    for i in range(1, d+1):
        cycle_list.append(i)
    cycle = S_d(tuple(cycle_list))
    # –≤—Å–µ —Å—Ç–µ–ø–µ–Ω–∏ —Ü–∏–∫–ª–∞
    subgroup_Sd = []
    for i in range(d):
        if i == 0:
            element = S_d.identity()  # g‚Å∞ = e
        else:
            element = cycle
            # —É–º–Ω–æ–∂–∞–µ–º —Ü–∏–∫–ª —Å–∞–º –Ω–∞ —Å–µ–±—è i-1 —Ä–∞–∑
            for j in range(1, i):
                element = element * cycle
        subgroup_Sd.append(element)
    print(f"–∏–∑–æ–º–æ—Ä—Ñ–Ω–∞—è –ø–æ–¥–≥—Ä—É–ø–ø–∞ S_{d}:")
    for i, element in enumerate(subgroup_Sd):
        print(f"  g^{i} = {element}")
    print()
    # –ø–æ–∫–∞–∑—ã–≤–∞–µ–º —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ
    print("—Å–æ–æ—Ç–≤–µ—Ç—Å–≤–∏–µ:")
    for i in range(d):
        print(f"  {elem_numb}^{i} = {subgroup_elements[i]}  ‚Üî  {subgroup_Sd[i]}")
    print(f"\n‚ü®{elem_numb}‚ü© –≤ Z_{m}* –∏–∑–æ–º–æ—Ä—Ñ–Ω–∞ C_{d} ‚äÇ S_{d}")
    print("—Ç–µ–æ—Ä–µ–º–∞ –ö—ç–ª–∏: –ª—é–±–∞—è –∫–æ–Ω–µ—á–Ω–∞—è –≥—Ä—É–ø–ø–∞ G –∏–∑–æ–º–æ—Ä—Ñ–Ω–∞ –ø–æ–¥–≥—Ä—É–ø–ø–µ —Å–∏–º–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–π –≥—Ä—É–ø–ø—ã S_{|G|}.")
    return {
        'm': m,
        't': elem_numb,
        'subgroup_order': d,
        'subgroup_elements_Zm': subgroup_elements,
        'subgroup_elements_Sd': subgroup_Sd,
        'isomorphism': f"C_{d} ‚äÇ S_{d}"
    }
isomorphism_of_cyclic_subgroup_Zm_star(N)
```

```python
import sage.all
from sage.rings.finite_rings.finite_field_constructor import GF

N = 501443 % 20

def find_polynomial_roots(N: int):
    F4 = GF(4, 'a')  # –ø–æ–ª–∏–Ω–æ–º –Ω–∞–¥ F‚ÇÑ —Å—Ç–µ–ø–µ–Ω–∏ 9
    a = F4.gen()
    coefficients_F4 = [] # –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã a_i = (i + N) mod 4
    for i in range(9):
        coeff_value = (i + N) % 4
        if coeff_value == 0:
            coefficients_F4.append(F4(0))
        elif coeff_value == 1:
            coefficients_F4.append(F4(1))
        elif coeff_value == 2:
            coefficients_F4.append(a)
        else:
            coefficients_F4.append(a + 1)
    poly_F4 = F4['x'](coefficients_F4 + [F4(1)])  # –ø–æ–ª–∏–Ω–æ–º: x‚Åπ + a‚Çàx‚Å∏ + ... + a‚ÇÄ
    roots_F4 = [] # –∏—â–µ–º –∫–æ—Ä–Ω–∏ –≤ F‚ÇÑ
    for element in F4:
        if poly_F4(element) == 0:
            roots_F4.append(element)
    print(f"F‚ÇÑ –ø–æ–ª–∏–Ω–æ–º: {poly_F4}")
    print(f"–ö–æ—Ä–Ω–∏ –≤ F‚ÇÑ: {roots_F4}")
    print(f"–ù–∞–π–¥–µ–Ω–æ –∫–æ—Ä–Ω–µ–π: {len(roots_F4)}")
    print() 
    F7 = GF(7) # –ø–æ–ª–∏–Ω–æ–º –Ω–∞–¥ F‚Çá —Å—Ç–µ–ø–µ–Ω–∏ 6
    coefficients_F7 = [] # –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã b_i = (i + N) mod 7
    for i in range(7):
        coeff_value = (i + N) % 7
        coefficients_F7.append(F7(coeff_value))
    poly_F7 = F7['x'](coefficients_F7) # –ø–æ–ª–∏–Ω–æ–º: b‚ÇÜx‚Å∂ + ... + b‚ÇÄ
    roots_F7 = [] # –∏—â–µ–º –∫–æ—Ä–Ω–∏ –≤ F‚Çá
    for element in F7:
        if poly_F7(element) == 0:
            roots_F7.append(element)
    print(f"F‚Çá –ø–æ–ª–∏–Ω–æ–º: {poly_F7}")
    print(f"–ö–æ—Ä–Ω–∏ –≤ F‚Çá: {roots_F7}")
    print(f"–ù–∞–π–¥–µ–Ω–æ –∫–æ—Ä–Ω–µ–π: {len(roots_F7)}")
    return roots_F4, roots_F7

roots4, roots7 = find_polynomial_roots(N)
```

```python
import sage.all
from sage.rings.finite_rings.finite_field_constructor import GF

N = 501443 % 20

def check_polynomials(N):
    print("–ø—Ä–æ–≤–µ—Ä–∫–∞ –ø–æ–ª–∏–Ω–æ–º–æ–≤ –Ω–∞ –ø—Ä–∏–≤–æ–¥–∏–º–æ—Å—Ç—å")
    print("\n1. –ø–æ–ª–∏–Ω–æ–º –Ω–∞–¥ F5:")
    F5 = GF(5) # –ø–æ–ª–∏–Ω–æ–º –Ω–∞–¥ F5
    R5 = F5['x']
    x = R5.gen()
    c_coeffs = [] # —Å—á–∏—Ç–∞–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã c_i = (i + N) % 5
    for i in range(5):
        value = (i + N) % 5
        c_coeffs.append(F5(value))
    # —Å—Ç—Ä–æ–∏–º –ø–æ–ª–∏–Ω–æ–º: x^5 + c4*x^4 + c3*x^3 + c2*x^2 + c1*x + c0
    poly5 = x**5 + c_coeffs[4]*x**4 + c_coeffs[3]*x**3 + c_coeffs[2]*x**2 + c_coeffs[1]*x + c_coeffs[0]
    print(f"–ø–æ–ª–∏–Ω–æ–º: {poly5}")
    if poly5.is_irreducible():  # –ø—Ä–æ–≤–µ—Ä—è–µ–º –ø—Ä–∏–≤–æ–¥–∏–º–æ—Å—Ç—å
        print("–Ω–µ–ø—Ä–∏–≤–æ–¥–∏–º")
    else:
        factors = poly5.factor()
        print("–ø—Ä–∏–≤–æ–¥–∏–º")
        print(f"—Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ: {factors}")
    roots5 = [] # –∏—â–µ–º –∫–æ—Ä–Ω–∏
    for element in F5:
        if poly5(element) == 0:
            roots5.append(element)
    print(f"–∫–æ—Ä–Ω–∏ –≤ F5: {roots5}")
    print("\n2. –ø–æ–ª–∏–Ω–æ–º –Ω–∞–¥ F9:")
    F9 = GF(9, 'a')  # –ø–æ–ª–∏–Ω–æ–º –Ω–∞–¥ F9 
    a = F9.gen()
    R9 = F9['x']
    x = R9.gen()
    d_coeffs = [] # —Å—á–∏—Ç–∞–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã d_i = (i + N) % 9
    for i in range(4):
        value = (i + N) % 9
        d_coeffs.append(F9(value))
    # —Å—Ç—Ä–æ–∏–º –ø–æ–ª–∏–Ω–æ–º: x^4 + d3*x^3 + d2*x^2 + d1*x + d0
    poly9 = x**4 + d_coeffs[3]*x**3 + d_coeffs[2]*x**2 + d_coeffs[1]*x + d_coeffs[0]
    print(f"–ø–æ–ª–∏–Ω–æ–º: {poly9}")
    if poly9.is_irreducible(): # –ø—Ä–æ–≤–µ—Ä—è–µ–º –ø—Ä–∏–≤–æ–¥–∏–º–æ—Å—Ç—å
        print("–Ω–µ–ø—Ä–∏–≤–æ–¥–∏–º")
    else:
        factors = poly9.factor()
        print("–ø—Ä–∏–≤–æ–¥–∏–º")
        print(f"—Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ: {factors}")
    roots9 = [] # –∏—â–µ–º –∫–æ—Ä–Ω–∏
    for element in F9:
        if poly9(element) == 0:
            roots9.append(element)
    print(f"–∫–æ—Ä–Ω–∏ –≤ F9: {roots9}")

check_polynomials(N)
```

```python
import sage.all
from sage.rings.finite_rings.finite_field_constructor import GF

N = 501443 % 20

def extended_gcd_poly(a, b):
    if b == 0:
        return (1, 0, a)
    x0, x1 = 1, 0
    y0, y1 = 0, 1
    r0, r1 = a, b
    while r1 != 0:
        q = r0 // r1
        r2 = r0 - q * r1
        x2 = x0 - q * x1
        y2 = y0 - q * y1
        r0, r1 = r1, r2
        x0, x1 = x1, x2
        y0, y1 = y1, y2
    return (x0, y0, r0)

def find_gcd_representation():
    print("–Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –ù–û–î –∏ –µ–≥–æ –ª–∏–Ω–µ–π–Ω–æ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è")
    F11 = GF(11)
    R11 = F11['x']
    x = R11.gen()
    r_coeffs = [] # –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã r_i = (i + N) % 11
    for i in range(8):
        value = (i + N) % 11
        r_coeffs.append(F11(value)) 
    s_coeffs = [] # –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã s_i = (i + N) % 11
    for i in range(4):
        value = (i + N) % 11
        s_coeffs.append(F11(value))
    f = r_coeffs[0] + r_coeffs[1]*x + r_coeffs[2]*x**2 + r_coeffs[3]*x**3 + r_coeffs[4]*x**4 + r_coeffs[5]*x**5 + r_coeffs[6]*x**6 + r_coeffs[7]*x**7
    g = s_coeffs[0] + s_coeffs[1]*x + s_coeffs[2]*x**2 + s_coeffs[3]*x**3
    print(f"f(x) = {f}")
    print(f"g(x) = {g}")
    gcd_fg = f.gcd(g) # –Ω–∞—Ö–æ–¥–∏–º –ù–û–î
    print(f"–ù–û–î(f,g) = {gcd_fg}")
    u, v, w = extended_gcd_poly(f, g) # –Ω–∞—Ö–æ–¥–∏–º –ª–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ
    inv_w = w**-1 # –Ω–æ—Ä–º–∞–ª–∏–∑—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
    u_normalized = u * inv_w
    v_normalized = v * inv_w
    print(f"u(x) = {u_normalized}")
    print(f"v(x) = {v_normalized}")
    check = u_normalized*f + v_normalized*g # –ø—Ä–æ–≤–µ—Ä–∫–∞
    print(f"u*f + v*g = {check}")

find_gcd_representation()
```

```python
import sage.all
from sage.rings.finite_rings.finite_field_constructor import GF

N = 501443 % 20

def find_inverse():
    print("–Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –æ–±—Ä–∞—Ç–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ f^{-1} mod g")
    F13 = GF(13)
    R13 = F13['x']
    x = R13.gen()
    s_coeffs = [] # –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã s_i = (i + N) % 11
    for i in range(3):
        value = (i + N) % 11
        s_coeffs.append(F13(value))
    # —Å—Ç—Ä–æ–∏–º –ø–æ–ª–∏–Ω–æ–º—ã
    f = s_coeffs[2]*x**2 + s_coeffs[1]*x + s_coeffs[0]
    g = x**8 + x**4 + x**3 + 6*x + 2
    print(f"f(x) = {f}")
    print(f"g(x) = {g}")
    gcd_val, h, _ = f.xgcd(g) # –Ω–∞—Ö–æ–¥–∏–º –æ–±—Ä–∞—Ç–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç
    if gcd_val == 1: # –ø—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ –æ–±—Ä–∞—Ç–Ω—ã–π —Å—É—â–µ—Å—Ç–≤—É–µ—Ç
        print(f"f^(-1) mod g = {h}")
        check = (f * h) % g # –ø—Ä–æ–≤–µ—Ä–∫–∞
        print(f"–ø—Ä–æ–≤–µ—Ä–∫–∞: f * h mod g = {check}")
    else:
        print("–æ–±—Ä–∞—Ç–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç")

find_inverse()
```

```python
import sage.all
from sage.rings.finite_rings.finite_field_constructor import GF

def generate_irreducible_polynomials(q: int, d: int) -> list:
    Fq = GF(q) # –∫–æ–Ω–µ—á–Ω–æ–µ –ø–æ–ª–µ F_q
    R = Fq['x'] # –∫–æ–ª—å—Ü–æ –ø–æ–ª–∏–Ω–æ–º–æ–≤ –Ω–∞–¥ F_q
    x = R.gen() # –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –¥–ª—è –≤–µ—Ä–Ω–æ–≥–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è 
    irreducible_polys = []
    # –ø–µ—Ä–µ–±–∏—Ä–∞–µ–º –≤—Å–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã
    from itertools import product
    for coeffs in product(Fq, repeat=d+1): # –¥–µ–∫–∞—Ä—Ç–æ–≤–æ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ, –≥–¥–µ –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç –¥–ª–∏–Ω–Ω–æ–π repeat
        if coeffs[d] != 0:  # —Å—Ç–∞—Ä—à–∏–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –Ω–µ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å 0
            # —Å—Ç—Ä–æ–∏–º –ø–æ–ª–∏–Ω–æ–º –æ–±—ã—á–Ω—ã–º —Ü–∏–∫–ª–æ–º
            poly = Fq(0)
            for i in range(d+1):
                poly = poly + coeffs[i] * (x**i)
            if poly.is_irreducible():
                irreducible_polys.append(poly)
    return irreducible_polys

test_cases = [(2,2), (2,3), (2,4), (3,2), (3,3), (5,2)]

for q, d in test_cases:
    print(f"\n–Ω–µ–ø—Ä–∏–≤–æ–¥–∏–º—ã–µ –ø–æ–ª–∏–Ω–æ–º—ã —Å—Ç–µ–ø–µ–Ω–∏ {d} –Ω–∞–¥ F_{q}:")
    polys = generate_irreducible_polynomials(q, d)
    for poly in polys:
        print(f"  {poly}")
    print(f"–≤—Å–µ–≥–æ: {len(polys)}")
```
